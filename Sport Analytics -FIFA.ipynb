The following columns need to be converted for further analysis:


Converting wage to float datatype

```python
for i in range(0,len(fifa)):
    v=fifa['Wage'][i]
    if(v==v):
        for j in v:
            v=v.replace('€','')
            if j=='M':
                v=v.replace('M','')
                v1=float(v)*1e6
            elif j=='K':
                v=v.replace('K','')
                v1=float(v)*1e3
        fifa['Wage'][i]=v1
fifa['Wage']=fifa['Wage'].astype(float)
```

```python
fifa['Wage'].dtypes
```

Converting value to float datatype

```python
for i in range(0,len(fifa)):
    v=fifa['Value'][i]
    if(v==v):
        for j in v:
            v=v.replace('€','')
            if j=='M':
                v=v.replace('M','')
                v1=float(v)*1e6
            elif j=='K':
                v=v.replace('K','')
                v1=float(v)*1e3
        fifa['Value'][i]=v1
fifa['Value']=fifa['Value'].astype(float)
```
